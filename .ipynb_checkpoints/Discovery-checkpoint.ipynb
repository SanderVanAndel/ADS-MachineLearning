# Importing dataset and libs

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

replays = pd.read_csv('../data/starcraft.csv')
```

```python
replays.head()
```

# Some plots

## LeagueIndex

```python
plt.figure(figsize=(16, 8));
labels = ('Bronze', 'Silver', 'Gold', 'Diamond', 'Master', 'GrandMaster', 'Prof1', 'Prof2')
values = replays.LeagueIndex
y_pos = np.arange(len(labels))

plt.bar(y_pos, replays['LeagueIndex'].value_counts().tolist(), align='center', alpha=1)
plt.xticks(y_pos, labels)

plt.title('Leagueindex spread')
plt.ylabel('Amount of replays')
```
